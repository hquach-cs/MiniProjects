```python
import pandas as pd 
import matplotlib.pyplot as plt

GDP = pd.read_csv("GDP.csv",thousands=',')
LE = pd.read_csv("LE.csv")
Data = {'Country':GDP['Country Name'],'GDP':GDP['2017'],'LE':LE['2017']}
Data = pd.DataFrame(Data,columns=['Country','GDP','LE'])
Data = Data[Data['GDP'] > 0.1e13]
Data = Data.dropna()
x = Data['LE'].to_numpy()
y = Data['GDP'].to_numpy()
plt.scatter(x,y)
plt.xlabel("Life Expectancy")
plt.ylabel("GDP")
```

```python

import sklearn.linear_model
import numpy as np

model = sklearn.linear_model.LinearRegression()
x = np.reshape(x,(-1,1))
model.fit(x,y)
plt.scatter(x,y)
plt.plot(x,model.predict(x),color='k')
```

```python
x_test = [[78]] # GDP when life expectancy is 78
print(model.predict(x_test)) # GDP = [1.36574113e+13]
```
